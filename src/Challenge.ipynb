```python
# import packages
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/electric_vehicles.csv", sep=";", index_col=0)
df['City'] = df['City'].str.replace('\d+ ', '')
```

```python
growth_rate_2021=(df["2021"]-df["2020"])/df["2020"]
growth_rate_2020=(df["2020"]-df["2019"])/df["2019"]
growth_rate_2019=(df["2019"]-df["2018"])/df["2018"]
growth_rate_2018=(df["2018"]-df["2017"])/df["2017"]
growth_rate_mean=(growth_rate_2021+growth_rate_2020+growth_rate_2019+growth_rate_2018)/4
df["2022_pred"] = round(df["2021"]*(1+growth_rate_mean))
df["2023_pred"] = round(df["2021"]*((1+growth_rate_mean)**2))
df["2024_pred"] = round(df["2021"]*((1+growth_rate_mean)**3))
df
```

```python
dv_ev_final = pd.read_excel("../data/excel.xlsx")
dv_ev_final
```

```python
df = pd.merge(df, dv_ev_final, on="City", how="right")
df
```

```python
df["recomm_2022"] = round(df["Count"]*(1+(df["2022_pred"]-df["2021"])/df["2021"]))
df["recomm_2023"] = round(df["recomm_2022"]*(1+(df["2023_pred"]-df["2022_pred"])/df["2022_pred"]))
df["recom_2024"] = round(df["recomm_2023"]*(1+(df["2024_pred"]-df["2023_pred"])/df["2023_pred"]))
df
```
